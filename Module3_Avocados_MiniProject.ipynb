```python
import pandas as pd
df = pd.read_csv("avocado.csv")
```

```python
df
```

```python
df.dropna()
```

```python
df.drop(["Unnamed: 0","Date","region"],axis = 1, inplace = True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
df["type"] = LabelEncoder().fit_transform(df["type"] )
```

```python
df
```

```python
features = len(df.columns)

y = df.iloc[:, 0:1].values
X = df.iloc[:,1:features].values
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

X_validate, X_test, y_validate, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsRegressor 
# Choose value of K 

K = 8
model = KNeighborsRegressor(n_neighbors=K)
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
scores = []
results = 0
best_score = 0

K = range(1,15)

for i in K:
    model = KNeighborsRegressor(n_neighbors = i)
    model.fit(X_train, y_train)
    results = model.score(X_test, y_test)
    scores.append(round(results,2))
    
    if results > best_score:
        best_score = results
        best_k = i
        bestmodel = model
        
print (scores)
print (best_k)
```

```python
import matplotlib.pyplot as plt
plt.plot(K, scores)
```

```python
accuracy = bestmodel.score(X_validate, y_validate)
print("The best model has an accuracy of:", round(accuracy, 2))
```

```python
from sklearn.linear_model import LinearRegression

modelLR = LinearRegression().fit(X_train, y_train.ravel())
score = modelLR.score(X_test, y_test)
print (score)
```

```python
from sklearn.ensemble import RandomForestRegressor

modelRF = RandomForestRegressor(criterion = "mse", max_leaf_nodes = 100).fit(X_train, y_train)
score = modelRF.score(X_test, y_test)
print (score)
```

```python
from sklearn.metrics import mean_squared_error

print ("Accuracy:", score)
print ("MSE+)
```
