```python
import pandas as pd
import numpy as np
df = pd.read_csv("KaggleV2-May-2016.csv")
```

```python
df.dropna()
```

```python
df_select = df [["Gender","Age","Scholarship","Hipertension","Diabetes", "Alcoholism", "Handcap", "SMS_received"]]
```

```python
df_select
```

```python
df_select

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder


scaler = StandardScaler()
encoder = OneHotEncoder()
label = LabelEncoder()

selected_X = df_select.iloc[:, 0:1].values
encoded_X = OneHotEncoder().fit_transform(selected_X).toarray ()
```

```python
selected_y = df_select.iloc[:,1:8].values
encoded_y = scaler.fit_transform(selected_y)
print(encoded_y)
```

```python
x = np.concatenate((encoded_X,encoded_y),axis=1)
```

```python
y = df['No-show'].values
y = label.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,
 test_size=0.2, random_state=0)
```

```python
x_validate, x_test, y_validate, y_test = train_test_split(x_test, y_test,
 test_size=0.5, random_state=0)
```

```python
print (len(x_train),len(x_test),len(x_validate))
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion = 'entropy').fit(x_train,y_train)
results = model.score(x_test,y_test)

results_validate = model.score(x_validate,y_validate)


print(results)
print(results_validate)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion = 'gini').fit(x_train,y_train)
results = model.score(x_test,y_test)

results_validate = model.score(x_validate,y_validate)


print(results)
print(results_validate)

from sklearn.metrics import confusion_matrix
predictions = model.predict(x_validate)
print("Confusion Matrix: \n", confusion_matrix(y_validate,predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier 
model = RandomForestClassifier(n_estimators=10, 
criterion='entropy')
model.fit(x_train, y_train)

results = model.score(x_test,y_test)

results_validate = model.score(x_validate,y_validate)


print(results)
print(results_validate)

from sklearn.metrics import confusion_matrix
predictions = model.predict(x_validate)
print("Confusion Matrix: \n", confusion_matrix(y_validate,predictions))
```
